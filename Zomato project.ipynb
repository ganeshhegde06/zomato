````python
##Sentiment Analysis
````

````python
from textblob import TextBlob

# Apply TextBlob sentiment analysis
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Create Sentiment column
merged_df["Sentiment"] = merged_df["Review"].apply(get_sentiment)

# Count sentiment distribution
sentiment_counts = merged_df["Sentiment"].value_counts()
sentiment_counts
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.    
Null Hypothesis (H): There is no significant difference in restaurant ratings between high-cost and low-cost restaurants.

Alternative Hypothesis (H): High-cost restaurants have significantly different ratings compared to low-cost restaurants.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
````

````python
import scipy.stats as stats

# 1. Testing the effect of Cost on Ratings (Independent T-test for High-cost vs. Low-cost restaurants)
median_cost = merged_df["Cost"].median()
high_cost_ratings = merged_df[merged_df["Cost"] > median_cost]["Rating"]
low_cost_ratings = merged_df[merged_df["Cost"] <= median_cost]["Rating"]

t_stat, p_value_cost = stats.ttest_ind(high_cost_ratings, low_cost_ratings, nan_policy='omit')
p_value_cost
````

````python
Effect of Cost on Ratings (T-test)

p-value: 3.26e-34 (very small)

Conclusion: Since the p-value is much lower than 0.05, we reject the null hypothesis. This means that cost does have a significant impact on restaurant ratings, indicating that higher-priced restaurants tend to receive different ratings compared to lower-priced ones.
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
````

````python
# Function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers for "Cost"
cost_outliers, cost_lower, cost_upper = detect_outliers_iqr(merged_df, "Cost")

# Detect outliers for "Rating" (should be within 0-5 range)
rating_outliers = merged_df[(merged_df["Rating"] < 0) | (merged_df["Rating"] > 5)]

# Apply outlier treatments
# Capping "Rating" to the valid range of 0-5
merged_df["Rating"] = merged_df["Rating"].clip(lower=0, upper=5)

# Capping "Cost" outliers to the upper and lower bounds
merged_df["Cost"] = merged_df["Cost"].clip(lower=cost_lower, upper=cost_upper)

# Check if outliers still exist
cost_outliers_after, _, _ = detect_outliers_iqr(merged_df, "Cost")
rating_outliers_after = merged_df[(merged_df["Rating"] < 0) | (merged_df["Rating"] > 5)]

# Output results
len(cost_outliers), len(cost_outliers_after), len(rating_outliers), len(rating_outliers_after)
````
